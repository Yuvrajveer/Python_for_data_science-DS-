```python
x= df['YearsExperience']
y= df['Salary']
```

Importing Stats_model for the Linear Regression

```python
import statsmodels.formula.api as smf
```

```python
model= smf.ols('y~x',data=df).fit()
model
```

```python
model.params
```

```python
model.pvalues
```

Predicting Salary_hike

```python
test=pd.DataFrame({
    'x':[11,21,33,41,17]
})
test
```

```python
model.predict(test)
```
