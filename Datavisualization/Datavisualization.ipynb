# Matplotlib HISTOGRAM

```python
df=pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/cars.csv')
df
```

```python
plt.hist(df['height'])
```

```python
df.hist()
plt.tight_layout()
```

```python
df['height'].skew()
```

```python
mean=df['height'].mean()
median=df['height'].median()
```

```python
df['height']
```

```python
plt.hist(df['height'] , edgecolor ='black')
plt.axvline(mean, color='green')
plt.axvline(median, color='pink')
plt.figure(dpi=1000)
```
